#Install Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Load Dataset

Use this dataset.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/exercise.csv"
df = pd.read_csv(url)

df.head()
```

This loads a sample dataset for analysis.

#Basic Data Exploration

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

##**Visualization 1 – Distribution Plot**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['pulse'], kde=True)
plt.title("Pulse Rate Distribution")
plt.show()
```

Shows how pulse rates are distributed among students.

##**Visualization 2 – Box Plot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='kind', y='pulse', data=df)
plt.title("Pulse Rate by Exercise Type")
plt.show()
```

Compare pulse rate across exercise categories.

##**Visualization 3 – Scatter Plot**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='pulse', y='time', data=df)
plt.title("Pulse vs Exercise Time")
plt.show()
```

Relationship between exercise time and pulse rate.

##**Correlation Heatmap**

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

This shows relationships between variables.
